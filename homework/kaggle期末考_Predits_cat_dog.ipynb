```
from tensorflow.python.keras import backend as K
from tensorflow.python.keras.models import load_model
from tensorflow.python.keras.preprocessing import image
import os
import numpy as np
```

```
from google.colab import drive
drive.mount('/content/drive')
```

```
net = load_model('/content/drive/My Drive/model-resnet50-final.h5') #載入模型
```

```
files = os.listdir('/content/drive/My Drive/test') #載入資料夾
files.sort()
```

```
images = [] #空串列
predits= [] #空串列
for file in files:
  img = image.load_img(os.path.join('/content/drive/My Drive/test', file), target_size=(224, 224)) #讀取圖片
  x = image.img_to_array(img) #圖片轉成數組
  x = np.expand_dims(x, axis = 0) #增加维度
  pred = net.predict(x)[0] #進行預測
  images.append(file.strip('.jpg')) #增加至空串列
  
  preds="{0:.10f}".format(pred[0]) #取預測值小數點10位
  predits.append(preds)#增加至空串列
  
  print(file) #列印圖片序列
  print(preds) #列印預測值
```

```
import pandas as pd
Dic= {"id":images,"Predicted":predits}
df = pd.DataFrame(Dic)
df.sort_index(inplace=True) #对DataFrame以index排序
df.to_csv('sample_submission.csv',index=None)
#df2 = pd.DataFrame({'id':np.arange(1,nb_test+1),'label':df['label']})
```
